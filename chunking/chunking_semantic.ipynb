```python
# ===== 셀 1: 시맨틱 청킹 =====
import json
from chonkie import SemanticChunker

# 문서 로드
DOC_PATH = "../docs/세종시교육청_자체감사규정.md"
with open(DOC_PATH, "r", encoding="utf-8") as f:
    text = f.read()

# 시맨틱 청커 (bge-m3 임베딩 기반)
chunker = SemanticChunker(
    embedding_model="BAAI/bge-m3",
    threshold=0.3,
    chunk_size=1024,
    min_sentences=2,
)

chunks = chunker.chunk(text)
print(f"시맨틱 청크 수: {len(chunks)}")
```

```python
# ===== 셀 2: JSONL 저장 =====
OUT_PATH = "semantic_chunks.jsonl"

with open(OUT_PATH, "w", encoding="utf-8") as f:
    for i, ch in enumerate(chunks, start=1):
        rec = {
            "chunk_id": i,
            "text": ch.text.strip(),
            "token_count": ch.token_count,
        }
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")

print(f"저장 완료: {OUT_PATH}")
```

```python
# ===== 셀 3: 결과 확인 =====
with open(OUT_PATH, "r", encoding="utf-8") as f:
    lines = f.readlines()

print(f"총 {len(lines)}개 청크\n")
for line in lines[:3]:
    rec = json.loads(line)
    print(f"--- chunk {rec['chunk_id']} (tokens: {rec['token_count']}) ---")
    print(rec["text"][:200], "...\n")
```
